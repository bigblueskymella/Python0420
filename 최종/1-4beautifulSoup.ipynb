```python
#🟠pip install requests
import requests # 웹사이트에 접속 🍋요청을 보내는 도구
from bs4 import BeautifulSoup

# 1. 네이버 메인 페이지 주소 #2️⃣구글닷컴으로 넣어보기
url = 'https://www.naver.com' 

# 2. 해당 주소로 "페이지 정보 좀 보내줘!" 요청하기
response = requests.get(url)

# 3. 받아온 복잡한 HTML 데이터를 BeautifulSoup으로 요리하기 좋게 변환 
# response. 은 안의 text
soup = BeautifulSoup(response.text, 'html.parser')

# 4. 네이버 로고 옆에 있는 서비스 이름들(메일, 카페, 블로그 등) 중 하나 찾아보기
# 실제 네이버 구조에 따라 달라질 수 있지만, 'span' 태그의 텍스트를 하나 가져와 봅니다.
print("네이버 페이지에서 가져온 텍스트:", soup.title.text)

# 🍋웹 크롤링의 뼈대
# 1. 접속 (requests)
# 2. 분석 (BeautifulSoup)
# 3. 추출 (find, title 등 태그 찾기)
```
