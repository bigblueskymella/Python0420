```python
#🟢BeautifulSoup 정적인 웹크롤링 // 직접 방문 requests 라이브러리의 도움
# pip install requests 🟢p276
import requests
import pandas as pd
from bs4 import BeautifulSoup
import time #시간 지연 도구

# 1. 수집한 모든 제목을 담을 빈 주머니(리스트) 준비
all_news_titles = []

# 2. 반복문(for)을 이용해 1페이지부터 3페이지까지 순회
# range(1, 4)는 1, 2, 3을 의미.
# 너무 빠르게 요청을 보내면 네이버 서버가 공격으로 오해 우리 IP 차단할 수
for page_num in range(1, 5):
    # f-string을 사용하여 🍋f{} URL의 page 번호를 자동 바꿈. IT/과학은 105 사회 뉴스는 102
    #네이버 뉴스 '사회' 섹션을 클릭 후 주소 확인 https://news.naver.com/main/main.naver?mode=LSD&mid=shm&sid1=102
    url = f'https://news.naver.com/main/main.naver?mode=LSD&mid=shm&sid1=102&date=%2000:00:00&page={page_num}'
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/110.0.0.0 Safari/537.36'}

    # 페이지 요청 및 HTML 파싱
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # 뉴스 제목 요소 찾기 (🟠클래스명)
    titles = soup.select('.sa_text_title')
    
    # 현재 페이지에서 찾은 제목들을 리스트에 추가
    for title in titles:
        clean_title = title.text.strip() #🔴앞뒤 공백 및 \n 제거
        all_news_titles.append(clean_title) #리스트(list)의 맨 마지막에 새로운 요소를 추가
    
    print(f"--- {page_num}페이지 수집 완료 (현재 총 {len(all_news_titles)}건) ---")
    time.sleep(1) #서버에 무리를 주지 않기 위해 1초씩 쉼.

# 3. 🟢★전체 리스트를 🍋데이터프레임(df_)으로 변환 ★
df_all_news = pd.DataFrame(all_news_titles)

# 4. 결과 확인 및 엑셀 저장 (185행)
print("\n" + "="*30) #==============================이거
print(f"최종 수집 완료! 총 {len(df_all_news)}건의 데이터가 확보되었습니다.")
print("="*30)

df_all_news.to_excel('naver_news_multi_page.xlsx', index=False)
print("\n--- naver_news_multi_page.xlsx 파일 저장 완료! ---")

print("\n--- 수집된 데이터의 상위 5개 항목 (head) ---")
# 기본 5개이며, 🍋괄호 안에 숫자 (예: .head(10)) 그만큼. 해보고 말.
print(df_all_news.head())
# 데이터의 전체 요약 정보 확인 (행/열 개수, 데이터 타입 등)
#print("\n--- 데이터프레임 요약 정보 (info) ---")
#print(df_all_news.info())
```

🟠깃허브 해당파일의 정답지 (수강생 안내용)

1번: User-Agent (컴퓨터가 아닌 사람임을 증명하는 헤더 정보)
2번: requests (서버에 요청을 보낼때)
3번: .sa_text_title (네이버 뉴스 제목 요소의 클래스 식별자)
4번: text (HTML 태그를 제외한 순수 텍스트만 가져오기)
5번: append (데이터를 리스트 주머니에 하나씩 쌓기)
6번: sleep (매너 있는 크롤링을 위한 시간 지연)
7번: DataFrame (Pandas의 핵심, 데이터를 표 구조로 만들기)

1. "여론의 규모"를 파악할 수 있습니다 (전수 조사)
