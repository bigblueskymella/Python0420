```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 난수10개 cumsum누적합계 인덱스 10개=0,10,20...90(0부터100까지 10단위)
s = pd.Series(np.random.standard_normal(10).cumsum(), index=np.arange(0, 100, 10))

plt.rcParams['font.family'] = 'Malgun Gothic'

s.plot(title="시간 경과에 따른 누적 변화량", marker='o') # 🟠 marker를 넣으면 지점이 더 잘 보입니다.
plt.xlabel("시간 (분)")
plt.ylabel("누적 수치")
plt.show()
```
