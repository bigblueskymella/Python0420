```python
import pandas as pd
import glob

# 1. 파일 목록 가져오기
file_list = glob.glob('월별_민원_*.xlsx')

# 2. 컬럼명을 통일하기 위한 매핑 사전(Dictionary) 만들기
# {'기존이름': '변경할이름'} 형식입니다. 2️⃣ 4줄
column_map = {
    '날짜': '일자', '접수일자': '일자', #'일시': '일자',
    '민원내용': '내용', '접수내용': '내용', #'상세내용': '내용',
    '처리상태': '상태' #, '상태': '상태', '진행단계': '상태'
}

all_data = []

for file in file_list:
    df = pd.read_excel(file)
    
    # 3. df.rename()으로 컬럼명 변경
    # columns 인자에 사전을 전달하면 해당되는 것만 골라 바뀜. 없는 컬럼명은 무시됨
    df.rename(columns=column_map, inplace=True) #2️⃣
    
    #🟢각 파일이 몇 줄인지 출력해보기 (디버깅용) 3️⃣
    # print(f"{file} 읽기 완료: {len(df)}행")

    all_data.append(df)

# 4. 표준화된 데이터프레임들을 하나로 합치기
total_df = pd.concat(all_data, ignore_index=True)

# 5. 필요한 컬럼만 최종 추출 (선택 사항)
total_df = total_df[['일자', '내용', '상태']]

# total_df.to_excel('통합_민원_데이터_최종1.xlsx', index=False) #1️⃣
# print("컬럼명 통합 🟠전!") #1️⃣

total_df.to_excel('통합_민원_데이터_최종2.xlsx', index=False) #2️⃣
print("컬럼명 표준화 및 통합 🟠완료!") #드래그 내려보면 지그재그로 됨 #2️⃣
```

```python
#컬럼명이 제각각인 엑셀 파일3개 만드는 코드

import pandas as pd
import numpy as np

# 실습용 가상 데이터 생성 함수
def create_sample_files():
    # 1월 데이터 (표준형)
    data1 = {
        '날짜': pd.date_range(start='2026-01-01', periods=100).strftime('%Y-%m-%d'),
        '민원내용': [f'1월 접수된 민원 사항 {i}번' for i in range(1, 101)],
        '처리상태': np.random.choice(['완료', '처리중', '보류'], 100),
        '담당부서': '민원지원팀'
    }
    pd.DataFrame(data1).to_excel('월별_민원_01월.xlsx', index=False)

    # 2월 데이터 (컬럼명 다름: 접수일자, 접수내용, 상태)
    data2 = {
        '접수일자': pd.date_range(start='2026-02-01', periods=100).strftime('%Y-%m-%d'),
        '접수내용': [f'2월에 발생한 민원 건 {i}번' for i in range(1, 101)],
        '상태': np.random.choice(['완료', '처리중', '보류'], 100),
        '비고': ''
    }
    pd.DataFrame(data2).to_excel('월별_민원_02월.xlsx', index=False)

    # 3월 데이터 (컬럼명 다름: 일시, 상세내용, 진행단계)
    data3 = {
        '일시': pd.date_range(start='2026-03-01', periods=100).strftime('%Y-%m-%d'),
        '상세내용': [f'3월 민원 신청 내역 {i}번' for i in range(1, 101)],
        '진행단계': np.random.choice(['완료', '처리중', '보류'], 100),
        '중요도': np.random.choice(['상', '중', '하'], 100)
    }
    pd.DataFrame(data3).to_excel('월별_민원_03월.xlsx', index=False)

    print("✅ 실습용 엑셀 파일 3개가 생성되었습니다!")

# 실행
create_sample_files()
```
