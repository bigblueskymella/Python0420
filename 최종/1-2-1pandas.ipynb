```python
import pandas as pd #🍋데이터프레임= 엑셀 같은 표
import matplotlib.pyplot as plt

# 1️⃣ 여러 개의 🟢리스트로 이루어진 데이터
items = ['랜덤 피규어', '캐릭터 키링', '장패드', '봉제 인형']
sales = [150, 120, 250, 180] # 판매 수량
prices = [15000, 8000, 25000, 35000]

# 2. 리스트를 합쳐서 🍋딕셔너리로, 다시 데이터프레임으로 변환
data = {
    '상품명': items,
    '판매량': sales,
    '가격': prices
}
df = pd.DataFrame(data) #데이터프레임(🍋df) 준비

# 3️⃣⭐ 새로운 열 '총매출' 생성
df['총매출'] = df['판매량'] * df['가격']
print(df) # 표 형태의 데이터 확인

# 2️⃣한글 깨짐 방지 설정 (Windows 기준, 필요시 설정)
plt.rcParams['font.family'] = 'Malgun Gothic'

# 1️⃣막대 그래프 1개 그리기
plt.bar(df['상품명'], df['판매량'], color='orange')
plt.title('상품별 판매 현황')
plt.xlabel('상품 종류')
plt.ylabel('판매량 (개)')

# 3️⃣두 개의 그래프를 나란히 그리기 (1행 2열)
# fig, axes = plt.subplots(1, 2, figsize=(15, 5))

# [왼쪽 그래프] 판매량 (개수)
# axes[0].bar(df['상품명'], df['판매량'], color='orange')
# axes[0].set_title('상품별 판매량 (수량 기준)')
# axes[0].set_ylabel('판매량 (개)')

# # [오른쪽 그래프] 총매출 (금액)
# axes[1].bar(df['상품명'], df['총매출'], color='skyblue')
# axes[1].set_title('상품별 총 매출액 (금액 기준)')
# axes[1].set_ylabel('매출액 (원)')
# plt.tight_layout()
plt.show()
```

그래프 두 개를 비교 - 재미있는 사실
판매량 1등: 그래프를 보면 '장패드'가 가장 많이 팔렸습니다. (박리다매)
매출액 1등: 하지만 실제 돈을 가장 많이 벌어다 준 효자 상품은 단가가 비싼 '봉제 인형'이나 '장패드'가 될 것입니다.

**"많이 파는 것이 중요한가, 비싸게 파는 것이 중요한가?"**에 대한 답을 이 새로운 열(총매출) 하나가 말해주는 것이죠.
