```python
#🟢유튜브 selenium 동적으로 
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By     
from selenium.webdriver.common.keys import Keys
import time
import pandas as pd

# 1. 브라우저 실행
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

# 2. 유튜브 특정 영상 페이지 접속 (예시 주요 이슈(예: 신규 사업 발표, 요금 인상 등 검색)
driver.get("https://www.youtube.com/watch?v=xFhEkerMuGc&list=RDxFhEkerMuGc&start_radio=1")
# driver.get("https://www.youtube.com/watch?v=6xbzX0LpRNg")
time.sleep(3) # 로딩 대기

# 3. 화면 아래로 스크롤 내리기 (댓글을 불러오기 위해)
# END 키를 눌러 페이지 맨 아래로 이동합니다.
body = driver.find_element(By.TAG_NAME, 'body') #By.ID CLASS_NAME TAG_NAME이 있음
for i in range(5): # 5번 스크롤 내리기
    body.send_keys(Keys.END) # 키보드의 END 키를 눌러라
    time.sleep(2) # 추가 데이터 로딩 대기

print("스크롤 완료! 이제 댓글을 수집할 준비가 되었습니다.")


# --------------------🔴2️⃣ '스크롤'이 반드시 선행되어야
# 4. 댓글 작성자와 내용 🍋찾아오기 find_elements
# 유튜브 댓글의 작성자 ID는 '#author-text', 내용은 '#content-text'라는 ID를 주로 사용
ids = driver.find_elements(By.CSS_SELECTOR, "#author-text")
comments = driver.find_elements(By.CSS_SELECTOR, "#content-text")

print(f"--- 총 {len(comments)}개의 댓글을 발견했습니다 ---")

# 5. 리스트에 담아서 🍋화면에 출력하기
youtube_data = []
for i in range(len(comments)):
    # .text로 글자만 추출하고, .strip()으로 앞뒤 공백 제거
    user_id = ids[i].text.strip()
    user_comment = comments[i].text.strip()
    
    print(f"{i+1}. [{user_id}] : {user_comment[:50]}...") # 너무 길어 50자만 출력
    youtube_data.append([user_id, user_comment])

# 6. Pandas 🍋데이터프레임으로 변환
df_youtube = pd.DataFrame(youtube_data, columns=['작성자', '댓글내용'])

# 🔴3️⃣ 유튜브 댓글을 엑셀로 저장. 어제 설치한 openpyxl 실행. 번호 열 안 만들기.
df_youtube.to_excel('youtube_comments.xlsx', index=False)
print("유튜브 댓글 엑셀 저장 완료!")

# 8. 브라우저 종료
driver.quit()
```
