```python
# 웹사이트에 접속 🍋요청을 보내는
import requests  
from bs4 import BeautifulSoup

# 1. 네이버 뉴스 메인 주소
url = 'https://news.naver.com/main/main.naver?mode=LSD&mid=shm&sid1=105' # IT/과학 뉴스 페이지

# 2. 브라우저인 척 (차단 방지) 서버가 "사람이 아니라 로봇이네?" 하고 차단 방지 위해 'User-Agent'를 넣습니다.
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/110.0.0.0 Safari/537.36'}

# 3. 페이지 정보 요청
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, 'html.parser')

# 4. 뉴스 제목이 들어있는 태그 찾기
# 네이버 뉴스 구조상 제목은 보통 'a' 태그의 'sh_text_headline' 클래스 등에 들어있습니다.
titles = soup.select('.sa_text_title') # 🔴🟠클래스 이름 앞에는 마침표(.)를 붙입니다.

print(f"--- 총 {len(titles)}개의 뉴스 제목을 찾았습니다 ---")

# 5. 반복문으로 1️⃣제목만 (번호붙여enumerate)하나씩 출력하기
# for i, title in enumerate(titles, 1): 2️⃣
# print(f"{i}. {title.text.strip()}")
for title in titles:
    print(f"{title.text.strip()}")

# 자바스크립트의 for...in은 **번호(index)**나 **이름(key)**을 가져오고, 
# 파이썬의 for...in은 **실제 데이터(value)**를 가져옴
```

```python
# 1. 필요한 도구(라이브러리) 가져오기
# import requests
# from bs4 import BeautifulSoup
import pandas as pd     #1-2-11️⃣pandas 두줄 가져오기
import matplotlib.pyplot as plt

# 2. 한글 폰트 설정 (그래프 깨짐 방지)
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False # 그래프-기호는 표준 폰트를 써서 그려라

# 3. 네이버 뉴스 데이터 수집 (크롤링) # 이전 단계이어서 해서 생략
# url = 'https://news.naver.com/main/main.naver?mode=LSD&mid=shm&sid1=105'
# headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/110.0.0.0 Safari/537.36'}

# response = requests.get(url, headers=headers)
# soup = BeautifulSoup(response.text, 'html.parser')
# titles = soup.select('.sa_text_title')

# 4. 수집한 데이터를 🍋'loaded_df'라는 이름의 표로 만들기 
# 리스트에 제목들을 담아 표로 만들기
title_list = [t.text.strip() for t in titles]
loaded_df = pd.DataFrame(title_list, columns=['오늘의 뉴스 제목'])
print(f"--- 총 {len(loaded_df)}개의 뉴스를 표로 변환했습니다 ---")
loaded_df.head() #1️⃣
# 5. 데이터 분석 및 시각화
# 실제 뉴스 제목에 포함된 단어 개수 세기 (AI, 삼성 등)
# .contains(k1): "그 글자들 속에 변수글 포함? case=False 대소문자구분x
k1, k2 = 'AI', '삼성' 
count1 = loaded_df['오늘의 뉴스 제목'].str.contains(k1, case=False).sum()
count2 = loaded_df['오늘의 뉴스 제목'].str.contains(k2, case=False).sum()

# # 🟠2️⃣ 현재 수집된 데이터를 '결과보고서.xlsx'로 저장 p247
# loaded_df.to_excel("news_report.xlsx", index=False)

# # 그래프 그리기
plt.figure(figsize=(8, 5))
plt.bar([f'{k1} 관련', f'{k2} 관련'], [count1, count2], color=['skyblue', 'orange'])
plt.title(f'실시간 뉴스 키워드 빈도 분석 ({k1} vs {k2})')
#plt.ylabel('기사 수', rotation=0, labelpad=20) #똑바로 세우기와 여백
plt.ylabel('기사 수')
# plt.xlabel('분석 키워드', labelpad=10) # x축 라벨도 너무 붙어있다면 pad
# plt.show()

# to_excel(): 파이썬 안의 데이터를 엑셀로 내보낼 때 씁니다.
# read_excel(): 엑셀에 있는 데이터를 파이썬(Pandas)으로 가져올 때 씁니다.
# 🟠pip install openpyxl 설치 명령어

# Pandas: 데이터를 표 형태로 관리하는 도구 (사무장)
# openpyxl: 데이터를 실제 엑셀 파일 규격으로 조립하는 도구 (현장 작업자)
```
