df.loc[0, '결제금액']

4. loc vs .iloc 한 줄 차이
.loc: 이름으로 찾기 (Location) - "성별", "결제금액", "금요일"
.iloc: 번호로 찾기 (Integer Location) - "0번 행", "1번 열

 결제금액이 2만원 이상인 사람들만 따로 모아서 표로 보여줘
target = loaded_df.loc[loaded_df['결제금액'] >= 20000]

🟢🟢
np.where는 "A 아니면 B"로 전체를 딱 나누고 싶을 때 빠르고 직관적. 
df.loc은 "특정 그룹만 콕 집어서 수정하자" 할 때 

```python
import pandas as pd

# 1. 🟢 결제금액 2만원 이상, 여자 행 찾기
# p208 🟢loc글자 iloc는 정수로만 색인을 취함
data = {
    '이름': ['김철수', '이영희', '박민수', '최지우', '정태양', '조하나'],
    '성별': ['남', '여', '남', '여', '남', '여'],
    '결제금액': [15000, 30000, 12000, 45000, 25000, 18000],
    '요일': ['금', '금', '토', '토', '금', '토']
}
loaded_df = pd.DataFrame(data)

print("--- [원본 데이터] ---")
print(loaded_df)
print("\n")

# 2. 결제금액 20,000원 이상인 행만 필터링 🟠.loc을 사용 True인 행만
filtered_df = loaded_df.loc[loaded_df['결제금액'] >= 20000]

print("--- [실습 1 결과: 2만원 이상] ---")
print(filtered_df)
print("\n")

# 3. 🟠두 가지 조건 (2만원 이상 & 성별이 '여') & 연산자와 괄호()가 핵심! 
# if 함수 대신, df.loc
target = loaded_df.loc[(loaded_df['결제금액'] >= 20000) & (loaded_df['성별'] == '여')]

# print("--- [실습 2 결과: 2만원 이상 여성] ---")
# print(target)

# 4. 🟢[str.contains 활용] '금요일'에 결제한 사람만 찾기
# 만약 요일 데이터가 카테고리형이라면 문자열 함수를 쓰기 위해 .astype(str)을 습관적으로 붙이기도
friday_df = loaded_df.loc[loaded_df['요일'].astype(str).str.contains('금')]

print("--- [실습 3 결과: 금요일 결제자] ---")
print(friday_df)

# 5. 🟢[응용] 이름에 '민'자가 들어가는 사람 찾기
name_filter = loaded_df.loc[loaded_df['이름'].str.contains('민')]

print("\n--- [실습 4 결과: 이름에 '민' 포함] ---")
print(name_filter)
```

```python
import pandas as pd
import numpy as np

# 🧲np.where랑 헷갈릴 수 있는 loc 
data = {
    '사업명': ['PJ-1', 'PJ-2', 'PJ-3', 'PJ-4', 'PJ-5'],
    '부서': ['운영팀', '기획팀', '운영팀', '인사팀', '기획팀'],
    '예산': [500, 10, 450, 5, 600],
    '상태': ['진행', '종료', '진행', '대기', '종료']
}
df = pd.DataFrame(data)

print("--- 수정 전 데이터 ---")
print(df)

# ---------------------------------------------------------
# 시나리오 1: 특정 값 찾아서 바꾸기 (엑셀의 '찾아바꾸기')
# [조건] '상태'가 '종료'인 데이터를 찾아 '완료'로 변경하세요.
# 문법: df.loc[행 조건, 열 이름] = 바꿀 값
# ---------------------------------------------------------
df.loc[df['상태'] == '종료', '상태'] = '완료'


# ---------------------------------------------------------
# 시나리오 2: 수치 비교를 통한 범주화 (엑셀의 IF 함수)
# [조건] '예산'이 100 미만인 소규모 사업의 '상태'를 '검토대상'으로 변경하세요.
# ---------------------------------------------------------
df.loc[df['예산'] < 100, '상태'] = '검토대상'


# ---------------------------------------------------------
# 시나리오 3: 여러 조건 만족 시 수정 (AND 조건)
# [조건] '부서'가 '운영팀'이고 '상태'가 '진행'인 사업에 '중요도'라는 새 컬럼을 만들고 '높음'을 입력하세요.
# ---------------------------------------------------------
df.loc[(df['부서'] == '운영팀') & (df['상태'] == '진행'), '중요도'] = '높음'


print("\n--- 수정 후 데이터 ---")
print(df)



# 👼loc대신 다른 필러링 query라는 것도 있다
# df.loc[(df['A'] > 1) & (df['B'] < 5)] 보다 
# df.query('A > 1 and B < 5')가 훨씬 직관적
```
