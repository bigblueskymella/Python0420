```python
#🟢구글 지도(Google Maps) 장소 리뷰 수집 selenium 동적으로 

from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
import pandas as pd

# 1. 브라우저 실행 및 구글 지도 접속
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
driver.get("https://www.google.com/maps")
time.sleep(3)

# 2. 장소 검색 (예: 서울 시청)
search_box = driver.find_element(By.CLASS_NAME, "UGojuc")
search_box.send_keys("북촌 한옥마을")
search_box.send_keys(Keys.ENTER)
time.sleep(5) # 결과 로딩 대기

# 3. '리뷰' 탭 클릭하기
# 구글 지도는 구조가 복잡해서 '리뷰'라는 글자가 적힌 버튼을 찾아 클릭합니다.
review_btn = driver.find_element(By.XPATH, '//button[contains(@aria-label, "리뷰")]')
review_btn.click()
time.sleep(3)

# 4. 리뷰 텍스트 크롤링 (상위 몇 개만)
# 리뷰 내용이 담긴 클래스명을 찾아 모든 리뷰를 가져옵니다.
reviews = driver.find_elements(By.CLASS_NAME, "wiI7pd") 

review_list = []
print(f"--- 수집된 리뷰 ({len(reviews)}건) ---")

for i, rev in enumerate(reviews[:10], 1): # 상위 10개만 출력
    content = rev.text.strip()
    print(f"{i}. {content[:50]}...")
    review_list.append(content)

# 5. Pandas로 저장 (3일차 연계)
df_reviews = pd.DataFrame(review_list, columns=['리뷰내용'])
# df_reviews.to_excel('google_maps_reviews.xlsx', index=False)
# print("엑셀 저장 완료!")
# driver.quit()

# 6. 데이터 정제 예시
# df_reviews.loc[df_reviews['리뷰내용'].str.contains('카페')]
# df_reviews['length'] = df_reviews['리뷰내용'].str.len()
# df_refined = df_reviews.loc[df_reviews['length'] >= 10]
# df_refined
```
