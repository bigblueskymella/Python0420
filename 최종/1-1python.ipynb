**주피터 노트북(Jupyter Notebook)**을 한마디로 정의한다면, 
**"데이터 분석의 디지털 연습장"**
🟢😁 강의용 비유: "한 권의 연구 노트(공책 - 낱장은 그냥 파이썬py)"

단순히 코딩만 하는 메모장이 아니라, **코드 + 실행 결과(표, 그래프) + 설명(글)**을 한 페이지에 모두 담을 수 있는 강력한 도구

🟢 주피터 노트북의 3대 핵심 특징
1. 주피터는 코드를 한 칸(Cell)씩 나눠서 실행할 수 있습니다.
    100줄짜리 코드 중 80번 줄에서 에러가 나면, 처음부터 다시 할 필요 없이 그 칸만 고쳐서
2. 코드를 실행하자마자 🟠그 바로 아래에 데이터 표(DataFrame)나 차트(Chart)가 나타납니다.
3. 마크다운 지원

.ipynb: 주피터 노트북 파일 (Interactive Python Notebook의 약자)
.py: 일반 파이썬 코드 파일 (한 번에 통째로 실행되는 문서)

[ ] 또는 ▶ 버튼이 있고, 칸마다 따로 실행할 수 있습니다.
```코드입력

````python
# 점수에 따른 등급 분류
score = 85

if score >= 90:
    grade = 'A'
elif score >= 80:
    grade = 'B'
elif score >= 70:
    grade = 'C'
else:
    grade = 'F'

print(f"점수: {score}, 등급: {grade}") # 결과: 점수: 85, 등급: B
print("점수:",score, ", 등급:", grade) # 결과: 점수: 85, 등급: B
````

````python
# 1. for문 : 리스트(배열)의 요소 합계 구하기
numbers = [10, 20, 30, 40, 50]
total = 0

for num in numbers:
    total += num #total = total + num 🟠들여쓰기 필수

print(f"결과 출력 🍋합계: {total}") 

# 2. for문 : 리스트/문자열 순회
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(fruit) # 🟠들여쓰기

# 3. range() 함수로 횟수 지정
for i in range(5): # 0부터 4까지 5번 반복
    print(i)
````

````python
# for + if 결합: 데이터 🟠필터링 (가장 중요)
# 50 이상인 숫자만 추출하여 리스트 만들기
data = [10, 60, 30, 80, 20, 90]
high_data = []

for item in data:
    if item >= 50:
        high_data.append(item) #덧붙이다: 리스트에 데이터 추가

print(f"50 이상 데이터: {high_data}") # 결과: [60, 80, 90]
````
