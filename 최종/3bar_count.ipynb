🟢 Countplot (개수 세기)"오늘 남자 손님은 몇 명, 여자 손님은 몇 명?   "데이터: [남, 여, 여, 남, 남]
  결과: 남자 3개, 여자 2개 → 막대 높이가 3과 2가 됨.
  특징: 그냥 줄 서 있는 머릿수만 세면 끝.
🟢 Barplot (평균 계산하기)"남자 손님이 쓴 돈의 평균과 여자 손님이 쓴 돈의 평균은 얼마?"
  데이터: [남: 1만원, 여: 2만원, 여: 3만원, 남: 2만원, 남: 3만원]결과: 남자 평균 2만원, 여자 평균 2.5만원 → 막대 높이가 2와 2.5가 됨.특징: '성별' 외에 '결제 금액'이라는 별도의 숫자가 있어야 계산 가능.

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# 1. 데이터 준비 (남성 3명, 여성 2명 / 각각 결제 금액 다름)
data = {
    '성별': ['남', '여', '여', '남', '남'],
    '결제금액': [10000, 20000, 30000, 20000, 30000]
}
df = pd.DataFrame(data)

# 2. 시각화
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.figure(figsize=(10, 4))

# [왼쪽] Countplot: 성별로 "몇 번" 나왔는지 🟢개수만 세기
plt.subplot(1, 2, 1)
sns.countplot(x='성별', data=df)
plt.title('Countplot: "몇 명 왔나?"')
# 결과: 남(3), 여(2)로 막대 높이가 정해짐

# [오른쪽] Barplot: 성별로 "결제금액의 🟢평균" 계산하기
plt.subplot(1, 2, 2)
sns.barplot(x='성별', y='결제금액', data=df)
plt.title('Barplot: "평균 얼마 썼나?"')
# 결과: 남(평균 2만), 여(평균 2.5만)로 막대 높이가 정해짐

plt.tight_layout()
plt.show()
```
