```python
#🟢selenium 동적인 웹크롤링 //대리인 보냄
# 1️⃣🟠🍋 pip install selenium webdriver-manager
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import time

# 1. 크롬 브라우저 실행(페이지 열기)
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

# 2. 네이버로 이동
driver.get("https://www.daum.net") # 2️⃣https://www.google.com로도 해보기
time.sleep(2) # 페이지가 로딩될 때까지 2초 🍋대기

# 3. 브라우저 제목 출력해보기
print("우리가 열리게 한 현재 페이지의 제목:", driver.title)

# 4. 종료 (이걸 실행하면 브라우저가 닫힙니다)
# driver.quit()



#-----------------------------------------------------------------
#🔴🍋2️⃣ 해당 사이트에서 검색입력후 엔터 
# q나 query로 되어있는 부분 검색
from selenium.webdriver.common.by import By # '어떤 기준(ID, 클래스 등)'으로 찾을지 결정하는 도구
from selenium.webdriver.common.keys import Keys # '엔터, 스페이스' 같은 키보드 입력을 담당하는 도구

# 브라우저 실행 (이미 켜져 있다면 위쪽 driver 변수를 그대로 써도 됩니다)
# driver.get("https://www.naver.com")

# 1. 검색창 찾기 (네이버 검색창의 ID는 'query' 다음은 'q' 구글 클래스 gLFyf)
# By.ID: 아이디로 찾기 By.CLASS_NAME: 클래스 By.CSS_SELECTOR: 주소(셀렉터)로 찾기
search_box = driver.find_element(By.ID, "q") 

# 2. 검색어 입력하고 엔터 치기
search_box.send_keys("파이썬으로 데이터 분석") #3️⃣다른 검색어 입력해보기
search_box.send_keys(Keys.RETURN) #검색창에서 엔터(Enter) 키를 눌러라

time.sleep(3) # 검색 결과가 나올 때까지 대기
print("검색 완료!")

#-----------------------------------------------------------------
#🔴🍋3️⃣ 검색 결과 타이틀 크롤링 #구글은 잘 안돼서.. 2번까지 다양하게.. 3은 다음, 네이버
# HTML에서 사용된 제목 클래스 이름 'news_tit'
titles = driver.find_elements(By.CLASS_NAME, "tit-g") #다음넷
# titles = driver.find_elements(By.TAG_NAME, "a")
print(f"--- 수집된 뉴스 제목 ({len(titles)}건) ---")

# 반복문을 사용하여 제목 하나씩 출력
for i, title in enumerate(titles, 1):
    print(f"{i}. {title.text}") # .text를 붙여야 HTML 태그 안의 '글자'만 가져옵니다.

# driver.quit()


#------------------------------------------------
#👼4️⃣ (번외)파일 저장 연습//상단 🟠1. import pandas as pd
# titles = driver.find_elements(By.CLASS_NAME, "tit-g") #다음넷
# print(f"--- 수집된 뉴스 제목 ({len(titles)}건) ---")

# news_list = [] #🟠2. 빈 바구니(리스트) 준비

# for i, title in enumerate(titles, 1):
#     print(f"{i}. {title.text}") 
#     news_list.append(title.text) #🟠3. 바구니에 하나씩 차곡차곡 담기

# df_all_news = pd.DataFrame(news_list, columns=['뉴스제목']) #news_list변수
# df_all_news.to_excel('save_practice.xlsx', index=False)
# print("엑셀저장완료🟠")
```
