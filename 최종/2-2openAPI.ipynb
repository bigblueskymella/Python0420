```python
import requests
import pandas as pd

# 1. 데이터를 요청할 주소 (보통 공공데이터포털에서 제공함)
# 실제 주소 예시: http://apis.data.go.kr/...
url = "https://jsonplaceholder.typicode.com/posts" # 연습용 무료 API 주소
# url = "https://api.github.com/repos/pandas-dev/pandas/issues" # 2️⃣p277

# 2. 서버에 데이터 요청하기
response = requests.get(url)
# response.raise_for_status() #2️⃣p277 상태확인
# print(response) #상태(오류)확인
# 3. 받은 데이터를 파이썬 형식(JSON)으로 바꾸기 🟢p277
data = response.json()

# 4. Pandas로 예쁘게 표 만들기🟢
df_api = pd.DataFrame(data)
#2️⃣p277(아래 두줄)
#  df_api = pd.DataFrame(data, columns=["number", "title", 
                                    #  "labels", "state"])

print("--- API로 가져온 데이터 상위 5줄 ---")
print(df_api.head()) #2️⃣🍋info, describe, tail, sample로 바꿔보기
```
