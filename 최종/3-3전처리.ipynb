```python
#🟢전처리: fillna와 groupby 시본이 추가됨

import pandas as pd # 데이터 표(DataFrame)를 다루는 라이브러리
import matplotlib.pyplot as plt # 기본 그래프를 그리는 도구
import seaborn as sns # 데이터 시각화 라이브러리 🟠1️⃣pip install seaborn 

# 1. 가상 데이터 생성 (전처리 실습)
data = {
    '부서': ['고객지원', '시설관리', '고객지원', '에너지사업', '시설관리', '고객지원'],
    '만족도': [85, 92, None, 78, 95, 88], # 결측치(None) 포함
    '접수건수': [120, 80, 150, 40, 90, 110]
}
df = pd.DataFrame(data) # 딕셔너리 형태의 데이터를 판다스 표(DataFrame) 형태로 변환

# 2. 전처리: 비어있는 만족도를 평균값으로 채우기
# '만족도' 컬럼의 빈 칸(NaN)을 찾아 전체 만족도의 평균값으로 채워넣음
# 🍋2️⃣mean()은 데이터들의 '산술 평균'을 구해주는 함수
df['만족도'] = df['만족도'].fillna(df['만족도'].mean())
# print("<빈칸 채운 값 보기> \n", df['만족도']) #🟢여기

# 3. 통계 분석: 부서별(3️⃣그룹화) 평균 만족도 계산
# '부서'별로 데이터를 🍋그룹 묶고, 🍋각 그룹의 '만족도' 평균을 계산하여 저장
dept_stats = df.groupby('부서')['만족도'].mean()
# print("<그룹화한 값 보기> \n", dept_stats) #🟢여기

# 4. 시각화 (한글 설정 필수)
plt.rcParams['font.family'] = 'Malgun Gothic' # 그래프 내 한글 깨짐 방지를 위해
plt.figure(figsize=(10, 6)) #도화지 가로, 세로 길이(단위: 인치) 

# 부서별 만족도 막대 그래프
# 🟠4️⃣시본없이: Pandas 객체(dept_stats)에서 바로 plot() 호출
# sns.barplot()으로 그림을 plt.title()이나 plt.savefig(저장)로 세부 마무리
dept_stats.plot(kind='bar', color=['skyblue', 'orange', 'green'], rot=0)

# 🍋x축은 부서 이름(index), 🍋y축은 계산된 평균값(values), 색상 테마는 'viridis'
# 🔴시본 바플롯과 카운트플롯
# viridis 대신 pastel이나 magma로 5️⃣바꿔서 //5️⃣이름 뒤에 _r 방향 반대 (예: viridis_r )
# sns.barplot(x=dept_stats.index, y=dept_stats.values, palette='viridis')

plt.title('부서별 고객 만족도 현황', fontsize=25) # 그래프 제목 설정 및 글자 크기 지정
plt.ylabel('평균 만족도 (점)') # y축의 이름(레이블) 설정
# plt.ylabel('평균 만족도 (점)', rotation=0, labelpad=40) 
# 회전과, 글자를 가로로 돌리면 축 숫자와 겹칠 수 있는데, 이때 숫자를 키워 축과의 간격을 벌려줍니다
plt.show()
```
